```python
import numpy as np
import pandas as pd
data_path = "../data/"
```

```python
df_state_abbrev = pd.read_csv(data_path + 'state_abbrev.csv')   
```

```python
df_bna = pd.read_csv(data_path + "bna.csv")
```

```python
df_comm_s = pd.read_csv(data_path + 'small_cities.csv', skiprows=5)
df_comm_m = pd.read_csv(data_path + 'medium_cities.csv', skiprows=5)
df_comm_l = pd.read_csv(data_path + 'large_cities.csv', skiprows=5)
tot_len   = len(df_comm_s) + len(df_comm_m) + len(df_comm_l)
df_comm   = pd.concat([df_comm_s, df_comm_m, df_comm_l]) 
df_comm.set_index(np.arange(tot_len))
```

```python
percent_list = []
number_list  = []
n_fail        = 0
for i in range(len(df_bna)):
    match_found   = False
    name1         = df_bna['city'][i]
    state1_abbrev = df_bna['state'][i]
    state1        = df_state_abbrev.loc[df_state_abbrev['abbrev']==state1_abbrev.upper()].values[0][0].lower()
    #print("looking for match for " + name1 + " in " + state1 + " in the BNA dataset")
    for j in range(len(df_comm)):
        name2         = df_comm.iloc[j]['City'].lower()
        if name1 in name2 and state1 in name2:
            match_found = True
            #print("matched to " + name2 + " in the commuting dataset\n")
            number_list.append(df_comm.iloc[j]['Number of workers who commuted by bicycle'])
            percent_list.append(df_comm.iloc[j]['Percentage of workers'])
            break
    if match_found == False:
        #print("no match found\n")
        number_list.append(None)
        percent_list.append(None)
        n_fail += 1
df_bna['number_of_bcs'] = number_list
df_bna['percent_bcs']   = percent_list
print("percent failed: " + str(n_fail / len(df_bna)))
```
